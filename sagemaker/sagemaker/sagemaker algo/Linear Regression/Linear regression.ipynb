# Train On Sagemaker using Linera Regression

```python
import boto3
import re
```

To train the Model four steps needs to be followed
1. Data should be present in S3
2. Mention S3 bucket name from where you need to upload the data
3. Mention the Instance type on which Model needs to be trained
4. Mention S3 location wherein output prediction needs to be stores


### Step 1:Upload the data from Local to AWS S3 

```python
bucketNM = 'ec2-east-sagemaker-demo'

TrainFile = r'LR/LRTrain.csv'
TestFile = r'LR/LRTest.csv'
Valfile = r'LR/LRVal.csv'
ModelFolder=r'LR/model/'
```

```python
s3ModelOutput= r's3://{0}/{1}'.format(bucketNM,ModelFolder)
s3Train = r's3://{0}/{1}'.format(bucketNM,TrainFile)
s3Test = r's3://{0}/{1}'.format(bucketNM,TestFile)
s3Val = r's3://{0}/{1}'.format(bucketNM,Valfile)
```

```python
s3Train
```

```python
with open('LRTrain.csv','rb') as f:
    boto3.Session().resource('s3').Bucket(bucketNM).Object(TrainFile).upload_fileobj(f)
```

```python
with open('LRTest.csv','rb') as f:
    boto3.Session().resource('s3').Bucket(bucketNM).Object(TestFile).upload_fileobj(f)
```

```python
with open('LRVal.csv','rb') as f:
    boto3.Session().resource('s3').Bucket(bucketNM).Object(Valfile).upload_fileobj(f)
```

### Download Built in Sagemaker Algorithm in sagemaker instance 

```python

import sagemaker
from sagemaker import get_execution_role
```

```python
sagemakerSess=sagemaker.Session()
```

```python
role=get_execution_role()
```

```python
role
```

https://docs.aws.amazon.com/sagemaker/latest/dg/sagemaker-algo-docker-registry-paths.html

linear-learner

```python
sagemakerSess.boto_region_name
```

```python
ECRdockercontainer=sagemaker.amazon.amazon_estimator.get_image_uri(sagemakerSess.boto_region_name,'linear-learner','latest')
```

```python
ECRdockercontainer
```

```python
instance_type='ml.m4.xlarge'#'local'
#
```

### Build Model

```python
LRModel=sagemaker.estimator.Estimator(image_name=ECRdockercontainer,
                                      role=role,
                                      train_instance_count=1,
                                        train_instance_type=instance_type,
                                        output_path=s3ModelOutput,
                                        sagemaker_session=sagemakerSess,
                                        base_job_name ='LR-Demo-v1'
                                        )
```

```python
len(dfTrain.columns)
```

```python
LRModel.set_hyperparameters(feature_dim='auto',
                           predictor_type='regressor',
                           normalize_data=False,
                           epochs=1

                           )
```

```python
#LRModel.set_hyperparameters()
```

```python
LRModel.hyperparameters()
```

### Training DATA and Test Data location

```python
trainConfig=sagemaker.session.s3_input(s3_data=s3Train,content_type='text/csv')
```

```python
trainConfig
```

```python
testConfig=sagemaker.session.s3_input(s3_data=s3Test,content_type='text/csv')
```

# Train Model

```python
LRModel.fit({'train': trainConfig,'validation':testConfig})
```

```python
# Deploy Model
```

```python
predictmodel=LRModel.deploy(initial_instance_count=1,instance_type='ml.m4.xlarge',endpoint_name='LR-Demo-v1')
```

```python
from sagemaker.predictor import csv_serializer, json_deserializer
```

```python


predictmodel.content_type = 'text/csv'
predictmodel.serializer = csv_serializer
predictmodel.deserializer = None
```
